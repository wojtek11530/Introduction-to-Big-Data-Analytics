```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# !!! based on https://towardsdatascience.com/web-scraping-nba-stats-4b4f8c525994

all_seasons_stats = []

beg_year = 1980
end_year = 2020

for year in range(beg_year, end_year):
    url = "https://www.basketball-reference.com/leagues/NBA_{}_per_game.html".format(year)
    # this is the HTML from the given URL
    html = urlopen(url)
    soup = BeautifulSoup(html)
    
    # getting headers of the table
    headers = [th.getText() for th in soup.findAll('tr', limit=1)[0].findAll('th')]
    headers = headers[1:]+['Year']

    rows = soup.find_all("tr", "full_table")
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
                for i in range(len(rows))]
    for single_player_stats in player_stats:
        single_player_stats.append(str(year))
        
    all_seasons_stats += player_stats
```

```python
stats_df = pd.DataFrame(all_seasons_stats, columns = headers)
stats_df.to_csv('nba_data.csv', index = False)
```
